```python
import matplotlib.pyplot as plt
import numpy as np


x = np.arange(0, 11)
y = x
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o')

plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Basic Line Graph for y = x")
plt.plot(x, y, color='pink')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
products = ["A", "B", "C", "D"]
sales = [40, 70, 30, 90]

# Bar chart
plt.figure(figsize=(8, 5))
plt.bar(products, sales, color="pink", edgecolor="black")

# Labels and title
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Sales of Products A, B, C, and D")

# Optional: Grid for readability
plt.grid(True, axis='y', linestyle="--", linewidth=0.7)

# Show chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(0, 10)
y = x**2
z = np.sin(x)
w = np.random.randint(1, 10, size=10)

# Create 22 subplot grid
plt.figure(figsize=(10, 8))

# Plot 1  Line Plot
plt.subplot(2, 2, 1)
plt.plot(x, y, color="pink")
plt.title("Line Plot (y = x)")
plt.xlabel("X")
plt.ylabel("Y")

# Plot 2  Bar Plot
plt.subplot(2, 2, 2)
plt.bar(x, w, color="pink", edgecolor="red")
plt.title("Bar Plot")
plt.xlabel("X")
plt.ylabel("Values")

# Plot 3  Scatter Plot
plt.subplot(2, 2, 3)
plt.scatter(x, z, color="red", marker='*')
plt.title("Scatter Plot (sin(x))")
plt.xlabel("X")
plt.ylabel("sin(X)")

# Plot 4  Histogram
plt.subplot(2, 2, 4)
plt.hist(w, color="pink", edgecolor="red")
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")

# Adjust layout
plt.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randn(200)

plt.hist(data, bins=20, edgecolor='red', color='pink')

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of 200 Normally Distributed Values")

plt.show()
```
